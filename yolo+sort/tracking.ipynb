```python
from sort.sort import Sort

from ultralytics import YOLO
import cv2
import os
import random
import numpy as np
import math
```

```python
video_input_path = os.path.join(".", "CityRoam.mp4")
video_out_path = os.path.join('.', 'out1.mp4')
cap = cv2.VideoCapture(video_input_path)
```

```python
ret, frame = cap.read()

cap_out = cv2.VideoWriter(video_out_path, cv2.VideoWriter_fourcc(*'MP4V'), cap.get(cv2.CAP_PROP_FPS),
                          (frame.shape[1], frame.shape[0]))
```

```python
tracker = Sort()
model = YOLO('best.pt')

colors = [(random.randint(0, 255), random.randint(0, 255), random.randint(0, 255)) for j in range(10)]

detection_threshold = 0.3
```

```python
while ret:
    results = model(frame, device="cpu", stream=True)
    detections = np.empty((0,5))
    for r in results:
        boxes = r.boxes
        for box in boxes:
            x1,y1,x2,y2 = box.xyxy[0]
            x1,y1,x2,y2 = int(x1),int(y1),int(x2),int(y2)

            cls = int(box.cls[0])

            conf = math.ceil(box.conf[0]*100)/100
            if conf > detection_threshold:
                currentArray = np.array([x1,y1,x2,y2,conf])
                detections = np.vstack((detections, currentArray))

    resultTracker = tracker.update(detections)

    for res in resultTracker:
        x1,y1,x2,y2,id = res
        x1,y1,x2,y2,id = int(x1),int(y1),int(x2),int(y2),int(id)
        w,h = x2-x1, y2-y1

        cv2.rectangle(frame, (int(x1), int(y1)), (int(x2), int(y2)), (colors[id % len(colors)]), 3)
        cv2.putText(frame, f'ID: {id}', (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2)

    cap_out.write(frame)
    ret, frame = cap.read()

cap.release()
cap_out.release()
cv2.destroyAllWindows()
```
